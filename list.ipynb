```python
#list: a list is a collection of items which is ordered and changeable. it allows duplicate members. lists are written with square brackets.
number_list = [1, 2, 3, 4, 5]
print(number_list)

#slicing: it allows you to access a portion of the list by specifying a range of indices. the syntax for slicing is list[start:stop:step].
print(number_list[1:4])  # it will print elements from index 1 to index 3 (4 is not included)
print(number_list[::2])  # it will print every second element of the list   



#methods in list


#append: it adds an element to the end of the list
number_list.append(6)
print(number_list)

#insert:  (position,value) it adds an element at the specified position
number_list.insert(2, 5)
print(number_list)


#remove:it removes the first occurrence of the specified value
number_list.remove(5,)
print(number_list)

```

```python
#extending list: extend() method is used to add all the items of an iterable (list, tuple, string etc.) to the end of the list. 
# It modifies the original list and does not return a new list.
a=[1,2]
a.extend([3,4])
print(a)

#pop(index): it is used to remove the element at the specified position and return it. If no index is specified, it removes and returns the last item in the list.
y=[10,20,30,40,50]
z=y.pop(2)
print(y)
print(z)
```

```python
#sort: it is used to sort the elements of a list in a specific order (ascending or descending). By default, it sorts in ascending order. 
# It modifies the original list and does not return a new list.

marks=[50,20,30,10,40]
marks.sort()#by default it is in ascending order
print(marks)

marks.sort(reverse=True)#descending order
print(marks)


#count: it is used to count the number of times a value is repeated in a list
marks=[50,20,30,10,40,50]
print(marks.count(50))
```
