```python
#multiplication table
N=int(input("enter a number:"))
for i in range(1,11):
    print(N,"*",i,"=",N*i)    
```

```python
#COUNT EVEN NUMBERS
count=0
N=int(input("enter a number:"))
for i in range(1,N+1):
    if i%2==0:
        count=count+1
print(count)

```

```python
#reverse counting
N=int(input("enter a number:"))
for i in range(N,0,-1):
    print(i,end=" ")
```

```python
#Square Series

N=int(input("enter a number:"))
for i in range(1,N+1):
    print(i**2,end=" ")
```

```python
#Count Digits
count=0
N=int(input("enter a number:"))
for i in str(N):
    count=count+1
print(count)
```
